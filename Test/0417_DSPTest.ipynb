```python
import thinkdsp
from thinkdsp import Sinusoid
from thinkdsp import normalize, unbias,TriangleSignal
import numpy as np
import matplotlib.pylab as pl
```

## Question: A sawtooth signal has a waveform that ramps up linearly from -1 to 1, then drops immediately to -1 and repeats. See http://en.wikipedia.org/wiki/Sawtooth_wave 

### (1) Write a class called 【Yourname_Sawtooth】 that extends Signal and provides evaluate to evaluate a sawtooth signal.  

```python
class TyTso_sawtooth(Sinusoid):
    
    def evaluate(self, ts):

        cycles = self.freq * ts + self.offset / np.pi / 2
        frac, _ = np.modf(cycles)
        ys = normalize(unbias(frac), self.amp)
        return ys
```

### (2) Generate a Sawtooth signal, with frequency= 1000 Hz, amplitude= 10,  plot it. 

```python
plotf = TyTso_sawtooth(1000,amp=10)
plotf.plot()
```

### (3) Compute the spectrum of the sawtooth wave as the above,  plot it. 

```python
plotw=TyTso_sawtooth(1000,amp=10).make_wave(framerate=40000)
plotw.make_spectrum().plot()
```
